```python
import pandas as pd
```

```python
products = driver.find_elements(By.CSS_SELECTOR,'div.-paxs.row._no-g._4cl-3cm-shs article.prd._fb.col.c-prd')
title_list = []
for product in products:
    product_name = product.find_element(By.CSS_SELECTOR,'div.info h3.name').text
    title_list.append(product_name)
    
```

```python
Product_Info = pd.DataFrame(title_list,columns=['Product Name'])
Product_Info
```

```python
URL_list = []
for product in products:
    product_url = product.find_element(By.CSS_SELECTOR,'a.core').get_attribute('href')
    URL_list.append(product_url)
    print(product_url)
```

```python
sleep(4)
Product_Info['Product URL'] = URL_list
Product_Info
```

```python
Product_Info=Product_Info.drop(columns='URL')
Product_Info
```

```python
Sale_price = []
for product in products:
    price = product.find_element(By.CSS_SELECTOR,'div.prc').text
    Sale_price.append(price)
Product_Info['Sale Price'] = Sale_price
Product_Info
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
product = driver.find_element(By.CSS_SELECTOR, "div.old")
print(product.text)
```

```python
old_price = []
for product in products:
    try:
        actual_price = product.find_element(By.CSS_SELECTOR,"div.old").text
        old_price.append(actual_price)
    except NoSuchElementException:
        old_price.append("N/A")
```

```python
Product_Info['Original Price'] = old_price
Product_Info
```

```python
Product_Info['Sale Price'] = Product_Info['Sale Price'].str.replace('EGP','', regex= False)
Product_Info['Original Price'] = Product_Info['Original Price'].str.replace('EGP','', regex= False)
Product_Info
```

```python
print(Product_Info['Sale Price'].dtype)
```

```python
Product_Info['Sale Price'] = Product_Info['Sale Price'].str.strip()
Product_Info['Original Price'] = Product_Info['Original Price'].str.strip()
Product_Info
```
